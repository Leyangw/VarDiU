```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
root_folder = '/home/leyang/projects/AnnealGen/inf_power_Gaussian_true_score'

# List subfolders
subfolders = [os.path.join(root_folder, d) for d in os.listdir(root_folder)
              if os.path.isdir(os.path.join(root_folder, d))]

valid_folders = []
for folder in subfolders:
    lists_folder = os.path.join(folder, 'lists')
    if (os.path.exists(os.path.join(lists_folder, 'time_list.json')) and
        os.path.exists(os.path.join(lists_folder, 'MMD_list.json')) and
        os.path.exists(os.path.join(lists_folder, 'logpd_list.json'))):
        valid_folders.append(lists_folder)

print(f"Found {len(valid_folders)} valid folders.")

time_lists = []
mmd_lists = []
logpd_lists = []

for lists_folder in valid_folders:
    with open(os.path.join(lists_folder, 'time_list.json')) as f:
        time_lists.append(np.array(json.load(f)))
    with open(os.path.join(lists_folder, 'MMD_list.json')) as f:
        mmd_lists.append(np.array(json.load(f)))
    with open(os.path.join(lists_folder, 'logpd_list.json')) as f:
        logpd_lists.append(np.array(json.load(f)))

time_array = np.array(time_lists)
mmd_array = np.array(mmd_lists)
logpd_array = np.array(logpd_lists)

avg_time0 = time_array.mean(axis=0)
print(len(avg_time0))

# First take log of MMD
mmd_array_safe = np.clip(mmd_array, 1e-12, None)  # avoid log(0)
log_mmd_array = np.log(mmd_array_safe)

# Now compute mean and std of log(MMD)
log_mmd_mean0 = log_mmd_array.mean(axis=0)
log_mmd_std0 = log_mmd_array.std(axis=0)

# logPD: divide by 10,000
logpd_array_scaled0 = logpd_array / 10000.0
logpd_mean0 = logpd_array_scaled0.mean(axis=0)
logpd_std0 = logpd_array_scaled0.std(axis=0)
print(logpd_mean0[-50:].mean(),log_mmd_mean0[-50:].mean())
print(logpd_std0[-50:].mean(),log_mmd_std0[-50:].mean())
```
